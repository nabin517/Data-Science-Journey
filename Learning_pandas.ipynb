# 4) Handling Missing Values

```python
df.isnull().sum() # provides the  no of null values in each column
```

```python
df.dropna(inplace=True) #removes rows that contain missing values (NaN) from the DataFrame.
```

```python
df['Type 2'].fillna(df['Type 2'].mode()[0]) # replacing the categorical column with its mode
```

```python
df['Type 2'].isnull().sum()# After filling with mode there is no missing values
```

# 5) Removing the Duplicate

```python
df.duplicated().sum() # provies the  no of duplicate data in my dataframe
```

```python
df.drop_duplicates(inplace=True) # removes the duplicate data
```

# 6) Renaming the Columns

```python
df.rename(columns={
    'HP':'Health',
    'Generation':'Gen'
},inplace=True) # renaming the columns 
```

# 7) Filtering The Rows

```python
df[df['Health']>120] # provides the row that have Health greater than 120
```

```python
df[(df['Health']>40) & (df['Name']=='Venusaur')] # gives the row that have health greater than 40 and the name is venusaur
```

```python
df.sort_values(by='Health',ascending=False) # sorting the data in descending order with Health
```

```python
df['Attacking_power']=df['Attack'].apply(lambda x: 'High' if x>100  else 'Low') # use of applying function
```
