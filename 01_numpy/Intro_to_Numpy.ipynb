# 2) Array Attributes

```python
arr1=np.array([[1,2,3],[4,5,6]])
```

```python
print('Shape',arr1.shape)
```

```python
print('Size',arr1.size)
```

```python
print('Dimension',arr1.ndim)
```

```python
print('Data Type',arr1.dtype)
```

```python
print(arr1.itemsize) # memory used by the one element
```

```python
print(arr1.nbytes) # total memory used ie nbytes=size*itemsize
```

# 3) Indexing and Slicing 

```python
print(arr1[0]) # printing the first element that is first row
```

```python
print(arr1[-1]) # printing the last element that is last row
```

```python
print(arr1[0][1]) # printing the second element inside the first element
```

```python
ar=np.array([10,20,30,40,50,60])
```

```python
print(ar[1:3])
```

```python
print(ar[:4])
```

```python
print(ar[::3]) # step slicing with the step 3
```

# slicing in 2D arrays

```python
print(arr1[0])
```

```python
print(arr1[:,1]) # print all the row but only the second column
```

```python
print(arr1[:,0]) # print all the row but only the first column
```

```python
print(arr1[0,1])
```

```python
# fancy indexing
print(ar[[1,3]])
```

```python
# boolean indexing 
print(ar[ar>25])
```

```python
# operations with scalars
print(c+2)
print(c*2)
```
